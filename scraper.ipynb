```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver import ActionChains
from selenium.webdriver.common.keys import Keys
import csv
import pandas as pd
import time
from selenium.common.exceptions import NoSuchElementException, TimeoutException
```

```python

def Telugana_link_route(driver, max_pages=2):
    State_Route_link = []
    Route_Telugana = []
    driver.implicitly_wait(30)
    wait = WebDriverWait(driver, 20)
    try:
        for page_number in range(1, max_pages + 1):
            print(f"Navigating to page {page_number}")

            elements = driver.find_elements(By.XPATH, "//a[@class='route']")
            for link_element in elements:
                d = link_element.get_attribute('href')
                State_Route_link.append(d)
                Route_Telugana.append(link_element.text)

            if page_number < max_pages:
                try:
                    next_page_button = wait.until(EC.element_to_be_clickable((By.XPATH, f"//div[@class='DC_117_pageTabs ' and text()='{page_number + 1}']")))
                    print("Next page button found, clicking...")
                    driver.execute_script("arguments[0].click();", next_page_button)
                    time.sleep(3)
                except (NoSuchElementException, TimeoutException) as e:
                    print(f"No more pages or error in locating next page button: {e}")
                    break
        return State_Route_link, Route_Telugana
    except (NoSuchElementException, TimeoutException) as e:
        print(f"Error in element path or initial page load: {e}")
        return [], []

# Example usage:
driver = webdriver.Chrome()
driver.get("https://www.redbus.in/online-booking/tsrtc/?utm_source=rtchometile")


Telugana_data = Telugana_link_route(driver, max_pages=2)
State_Route_link, Route_Telugana = Telugana_data
driver.quit()


for i in range(len(State_Route_link)):
    print(f"Link: {State_Route_link[i]}, Route: {Route_Telugana[i]}")
```

```python
data = {
    "Route_name": Route_Telugana,
    "State_Route_link": State_Route_link
}

# Create DataFrame
df = pd.DataFrame(data)

# Save to CSV
df.to_csv('C:/Users/sinwa/projects/RED_BUS_project/csv_files/Telugana.csv', index=False)


print("CSV file saved successfully.")
```
